```python
x = np.random.randint(50,200,20)
y = np.sin(x)
```

```python
fig,ax = plt.subplots(2,2,figsize=[10,10])

ax[0,0].plot(x,y,'--o')
ax[0,0].set_xlabel('A',fontsize=15)
ax[0,0].set_ylabel('B',fontsize=15)
ax[0,0].set_title('Line plot')
```

```python
fig,ax = plt.subplots(2,2,figsize=[10,10])

ax[0,0].scatter(x,y)
ax[0,0].set_xlabel('A',fontsize=15)
ax[0,0].set_ylabel('B',fontsize=15)
ax[0,0].set_title('Scatter plot',fontsize=15)
```

```python
fig,ax = plt.subplots(2,2,figsize=[10,10])

ax[0,0].barh(x,y)
ax[0,0].set_xlabel('A',fontsize=15)
ax[0,0].set_ylabel('B',fontsize=15)
ax[0,0].set_title('Barh plot',fontsize=15)
plt.show()
```

```python
fig,ax = plt.subplots(2,2,figsize=[10,10])

ax[0,0].hist(x,bins=8,alpha=0.8)
ax[0,0].set_xlabel('A',fontsize=15)
ax[0,0].set_ylabel('B',fontsize=15)
ax[0,0].set_title('Hist plot',fontsize=15)
#plt.show()
```

```python
states = ['MH','GJ','AP','GA','KA','UP']
```

```python
fig,ax = plt.subplots(2,2,figsize=[10,10])

ax[0,0].pie([750,450,589,256,354,582],labels=states,autopct='%.2f%%')
ax[0,0].set_xlabel('A',fontsize=15)
ax[0,0].set_ylabel('B',fontsize=15)
ax[0,0].set_title('Pie plot',fontsize=15)
plt.show()
```

```python
fig,ax=plt.subplots(3,2,figsize=[10,10])
ax[0,0].plot([np.linspace(2,10,50),np.sin(np.linspace(2,10,50))],'--o',color='m')
ax[0,0].set_xlabel('A',fontsize=15)
ax[0,0].set_ylabel('B',fontsize=15)
ax[0,0].set_title('Line Plot')

ax[0,1].bar(x,y)
ax[0,1].set_xlabel("AA",fontsize=15)
ax[0,1].set_ylabel("BB",fontsize=15)
ax[0,1].set_title('Bar Plot')

ax[1,0].hist(x,bins=[50,70,90,120,140,160,180,200],rwidth=0.96)
ax[1,0].set_xlabel('AAA',fontsize=15)
ax[1,0].set_ylabel('BBB',fontsize=15)
ax[1,0].set_title('HIst Plot')

ax[1,1].scatter(x,y)
ax[1,1].set_xlabel('AAAA',fontsize=15)
ax[1,1].set_ylabel('BBBB',fontsize=15)
ax[1,1].set_title('Scatter Plot')


ax[2,0].pie([750,542,652,125,265,489],labels=states,autopct='%.f%%')
ax[2,0].set_title('Pie Chart')
plt.suptitle('Plots are Line, Bar, Hist , Scatter and Pie',fontsize=25)
plt.tight_layout()
```
