```python
overs =[1,2,3,4,5,6]
runs =[13,20,25,36,11,10]
```

```python
plt.plot(overs,runs)
```

#to create cumulative of runs

```python
runs
```

```python
import numpy as np
```

```python
run_cum = np.array(runs).cumsum()
run_cum
```

```python
plt.plot(overs,run_cum)
```

```python
plt.plot(overs,run_cum,linewidth=10,color='yellow',label= 'CSK')
plt.xlabel('overs')
plt.ylabel('run_cum')
plt.title('CSK runs for 1 - 6 overs')
plt.legend()
```

```python
runs_RR = [5,23,22,17,30,36]
```

```python
runcum_RR = np.array(runs_RR).cumsum()
runcum_RR
```

```python
plt.plot(overs,runcum_RR,linewidth=10,color='blue',label='RR')
plt.xlabel('overs')
plt.ylabel('runcumulative_RR')
plt.legend()
plt.title('RR for 1-6 overs')
```

```python
plt.plot(overs,runcum_RR,linewidth=10,color='blue',label='RR')
plt.xlabel('overs')
plt.ylabel('runcumulative_RR')
plt.title('RR for 1-6 overs')
plt.legend()
```

```python
plt.plot(overs,runcum_RR,color='blue',linewidth=10,label='RR')
plt.plot(overs,run_cum,color='yellow',linewidth=5,label='CSK')
plt.xlabel('OVERS')
plt.ylabel('Cumulative Runs of both teams with diff. colors')
plt.title('CSK VS RR 1-6 overs comparison')
plt.legend()
```
